```python
import pandas as pd
import numpy as np
phone = pd.read_csv("https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/cell_phones.csv")
phone.head()
```

```python
population = pd.read_csv('https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/population1.csv')
population.head()
```

```python
geo_code = pd.read_csv('https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--entities--geo--country.csv')
geo_code.head()
```

```python
pop_phone = pd.merge(phone,population,how='inner',on=['Country','year'])
pop_phone.head()
```

```python
print(pop_phone.shape)
print()
pop_phone.info()
```

```python
geo_code = geo_code.rename({'country':'geo','name':'Country'},axis=1)
geo_code.head()
```

```python
geo_coun = geo_code.loc[:,['geo','Country']]
geo_coun.head()
```

```python
combined = pd.merge(pop_phone,geo_coun,how='inner',on='Country')
print(combined.shape)
print()
combined.head()
```

```python
combined.population.sum()
```

```python
combined['phone_per_person'] = combined.cell_phones_total / combined.population
combined.head()
```

```python
combined[(combined.Country == 'United States') & (combined.year == 2017)]
```

```python
combined['geo'] = combined['geo'].str.upper()
```
